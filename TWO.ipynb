<font size="3"> Filtering the data frame

```python
# keep only columns that include venue name, and anything that is associated with location
filtered_columns = ['name', 'categories'] + [col for col in dataframe.columns if col.startswith('location.')] + ['id']
dataframe_filtered = dataframe.loc[:, filtered_columns]

def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']

    
    
# filter the category for each row
dataframe_filtered['categories'] = dataframe_filtered.apply(get_category_type, axis=1)

# clean column names by keeping only last term
dataframe_filtered.columns = [column.split('.')[-1] for column in dataframe_filtered.columns]

dataframe_filtered
```

```python

df=dataframe_filtered.drop(columns=['address','cc','country','labeledLatLngs','postalCode','state','crossStreet'])

df
```

<font size="3">Now we are going to cluster these data and map it to see what the model yields

```python

# Matplotlib and associated plotting modules
import matplotlib.cm as cm
import matplotlib.colors as colors

# import k-means from clustering stage
from sklearn.cluster import KMeans



print('Libraries imported.')

```

```python
mapt = folium.Map(location=[25.809967, -80.311378], zoom_start=11)

# add markers to map
for lat, lng, name, in zip(df1['lat'], df1['lng'], df1['name']):
    label = '{}'.format(name)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(mapt)  
    
mapt
```

<font size=4> Now we are going to start clustering based on the data previously obtained:

```python
df2=df1.drop(columns=['formattedAddress','city','categories','id','rating','name'])
df2
```

```python
# set number of clusters
kclusters = 4



# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(df2)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:] 
```

```python
df3=df1.drop(columns=['formattedAddress','city','categories','id','rating',])
df3['cluster']=kmeans.labels_
df3.sort_values(by=['distance'])
```

```python
# create map
map_clusters = folium.Map(location=[25.809967, -80.311378], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, name, cluster in zip(df3['lat'], df3['lng'], df3['name'], df3['cluster']):
    label = '{},{}'.format(name,cluster)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
    [lat, lon],
    radius=5,
    popup=label,
    color=rainbow[cluster-1],
    fill=True,
    fill_color=rainbow[cluster-1],
    fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
